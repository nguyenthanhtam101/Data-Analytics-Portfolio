```python
import pandas as pd
file_path = '/content/sample_data/superstore.csv'
df = pd.read_csv(file_path)
```

```python
# 1. In ra thông tin tổng quan của dữ liệu
print("--- Thông tin ban đầu của dữ liệu ---")
df.info()

# 2. In ra 5 dòng đầu tiên
print("\n--- 5 dòng dữ liệu đầu tiên---")
df.head()
```

```python
#Tranform Order Date and Ship Date to datetime
df['Order.Date'] =  pd.to_datetime(df['Order.Date'])
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])
#Creat new features
df['Profit Margin'] = df['Profit']/ df['Sales']
df['Processing Time'] = (df['Ship.Date'] - df['Order.Date']).dt.days
```

```python
#Infomation was cleaned
df.info()
```

```python
df.to_csv('Superstore_Cleaned.csv', index=False)
```
